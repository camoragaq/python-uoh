##Random Walk

En esta sección usaremos Python para generar datos para un paseo aleatorio y luego usaremos matplotlib para crear una representación visualmente atractiva de los datos generados. 
    
    Un paseo aleatorio es un camino que NO tiene una dirección clara, sino que 
    está determinado por una serie de decisiones aleatorias (al azar), cada 
    una de las cuales se deja totalmente al azar. Se puede imaginar un paseo 
    aleatorio como el camino que seguiría una hormiga si perdiera la cabeza y 
    diera cada paso en una dirección aleatoria.
    
    Los paseos aleatorios tienen aplicaciones prácticas en la naturaleza, la 
    física, la biología la química y la economía. Por ejemplo, un grano de 
    polen que flota en una gota de agua se mueve por la superficie de esta 
    porque las moléculas de agua lo empujan constantemente. El movimiento 
    molecular en una gota de agua es aleatorio, por lo que el camino que traza 
    un grano de polen en la superficie es al azar. Asi, el código que vamos a 
    escribir puede modelar muchas situaciones del mundo real.

```python
#vamos a importar la siguientes librerías
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
# Creamos los datos a utilizar
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)#definimos 6 grupos de ejes con 500 elementos cada uno, y cumsum() retorna la suma cumulativa de los elementos de un eje dado

```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

Otra forma usando clases:

```python
from random import choice #selecciona un elemento aleatorio  

class RandomWalk():
    #Clase para generar camino aleatorio
    def __init__(self, num_points=5000): 
         #Inicializa atributos de del camino.
         self.num_points = num_points
         self.x_values = [0] 
         self.y_values = [0]


    def fill_walk(self):
        while len(self.x_values) < self.num_points:
           
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            if x_step == 0 and y_step == 0:
                continue
            
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x) 
            self.y_values.append(next_y)
            
rw = RandomWalk()
rw.fill_walk()
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, marker='o', s=5) 

plt.show() 
```
