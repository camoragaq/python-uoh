4. Dibuje un gráfico que compare peso y altura en 4 grupos distintos, generando valores random, usando las librerías MAtplotlib y NumPy.

```python
import matplotlib.pyplot as plt
import numpy as np 
rng = np.random.RandomState(0)
weight=np.linspace(100, 10, 500)
height=np.cumsum(rng.randn(500, 10), 0)
plt.plot(weight, height, marker='*')
plt.xlabel('weight', fontsize=16)
plt.ylabel('height', fontsize=16)
plt.title('Group wise Weight vs Height scatter plot',fontsize=20)
plt.show()
```
