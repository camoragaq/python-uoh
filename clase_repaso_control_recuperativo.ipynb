1. Escriba una función que retorne el mayor de 3 números. 

```python
def max2( x, y ):
    if x > y:
        return x
    return y
def max3( x, y, z ):
    return max2( x, max2( y, z ) )


print(max3(3, 6, -5))
```

2. Escriba una función que multiplique todos los números dado una lista.

```python
def multiplica(numeros):  
    total = 1
    for x in numeros:
        total *= x  
    return total  

print(multiplica((8, 2, 3, -1, 7)))
```

4. Escriba un programa en Python que verifique si un triángulo es equilátero, isósceles o escaleno.

```python
print("Ingrese los lados del triangulo: ")
x = int(input("x: "))
y = int(input("y: "))
z = int(input("z: "))

if x == y == z:
	print("Triangulo equilatero")
elif x==y or y==z or z==x:
	print("Triangulo isosceles")
else:
	print("Triangulo escaleno")
```

6. Escriba una función que cuente el número de palabras dentro de un archivo.

```python
def cuenta_palabras():
    file = open("texto.txt","r")
    contador = 0
    data = file.read()
    palabras = data.split("--")#separara la palabras y las guardara en una lista
    for i in palabras:
        contador += 1
    print("Total",contador)
    file.close()

cuenta_palabras()
```

7. Escriba una función que lea un archivo y muestre las palabras de largo mayor a 4.

```python
def muestra_palabras():
    file = open("texto.txt","r")
    data = file.read()#leemos el archivo 
    lista = data.split("--")#y aca lo separamos por espacios
    for i in lista:
        if len(i) > 4:
            print(i, end=" ")
    file.close()

muestra_palabras()
```

```python
ciudades=["rancagua","machali"]

for i in ciudades:#recorremos las ciudades primero y sumaremos cada ciudad en la lista
  sueldos=0 
  file=open("texto.txt","r")#leemos el archivo por cada ciudad para sumar cada linea
  for line in file:#aca recorremos las lineas
    a=line.split("--")
    if a[2]==i :
      sueldos=sueldos+int(a[3])
  salida=open(i+".txt","w")#imprimimos el archivo antes de pasar a la siguiente ciudad
  salida.write(str(sueldos))

```
